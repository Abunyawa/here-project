```python
import pandas as pd
import geopandas as gpd
import requests
from bs4 import BeautifulSoup
import json
```

```python
resp = requests.get("https://avoska.ru/api/get_shops.php?map=1").json()
resp
```

```python

items = []

for item in resp['features']:
    inToDF = {
    'id':item['id'],
    'latitude':item['geometry']['coordinates'][0],
    'longitude':item['geometry']['coordinates'][1],
    'address':item['properties']['hintContent']
    }
    items.append(inToDF)

df = pd.DataFrame(items)
df
```

```python
df.to_csv('../res/avoska_shops.csv',index=None);
```
